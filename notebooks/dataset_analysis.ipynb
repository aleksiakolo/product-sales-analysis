```r
library(dplyr)

clean_tables <- map(clean_tables, distinct)
```

```r
map(clean_tables, glimpse)
```

```r
# make returned in returns a logical variable
clean_tables$returns <- clean_tables$Returns %>%
  mutate(
    returned = returned == "Yes"
  )
```

```r
region_levels <- sort(unique(clean_tables$Orders$region))
market_levels <- sort(unique(clean_tables$Orders$market))
```

```r
# make categorical variables when necessary in orders
clean_tables$Orders <- clean_tables$Orders %>%
  mutate(
    ship_mode      = factor(ship_mode),
    segment        = factor(segment),
    order_priority = factor(order_priority, 
                             levels = c("Low", "Medium", "High", "Critical"),
                             ordered = TRUE),
    category       = factor(category),
    sub_category   = factor(sub_category),
    country        = factor(country),
    state          = factor(state)
  )
```

```r
clean_tables$Orders <- clean_tables$Orders %>%
  mutate(region = factor(region, levels = region_levels))

clean_tables$People <- clean_tables$People %>%
  mutate(region = factor(region, levels = region_levels))
```

```r
clean_tables$Orders <- clean_tables$Orders %>%
  mutate(market = factor(market, levels = market_levels))

clean_tables$Returns <- clean_tables$Returns %>%
  mutate(market = factor(market, levels = market_levels))
```

```r
clean_tables$Orders <- clean_tables$Orders %>%
  mutate(
    sales         = as.numeric(sales),
    quantity      = as.integer(quantity),
    discount      = as.numeric(discount),
    profit        = as.numeric(profit),
    shipping_cost = as.numeric(shipping_cost)
  )
clean_tables$Orders <- clean_tables$Orders %>%
  mutate(
    order_date = as.Date(order_date),
    ship_date  = as.Date(ship_date)
  )
```

```r
identical(
  levels(clean_tables$Orders$region),
  levels(clean_tables$People$region)
)
# TRUE
identical(
  levels(clean_tables$Orders$market),
  levels(clean_tables$Returns$market)
)
# TRUE
```

```r
output_dir <- "../data/processed"
dir.create(output_dir, recursive = TRUE, showWarnings = FALSE)

purrr::iwalk(
  clean_tables,
  ~ readr::write_csv(
      .x,
      file.path(output_dir, paste0(.y, ".csv"))
    )
)
```
